```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# simple_second_df = pd.read_pickle('tmp_germany_migration_0914.pkl')
simple_second_df = pd.read_pickle('tmp_US_migration_0914.pkl')
# country = 'Germany'
country = 'UnitedStates'
germany = simple_second_df[simple_second_df['pre_country'] == country]
print(len(germany))
# i8n = {}
all_ = {}

performance = {}
education = {}
start_org = {}
career = {}
settlement = {}
creation = {}
assembly = {}
military = {}

for i in range(1900, 2001):
    performance[i] = []
    education[i] = []
    start_org[i] = []
    career[i] = []
    settlement[i] = []
    creation[i] = []
    assembly[i] = []
    military[i] = []
    all_[i] = []


for idx, row in germany.iterrows():
    time = int(row['time'])

    if row['loc_country'] != country:
        if row['label'] == 'give birth':
            performance[time].append(row['name'])
        if row['label'] == 'education':
            education[time].append(row['name'])
        if row['label'] == 'start org':
            start_org[time].append(row['name'])
        if row['label'] == 'movement':
            career[time].append(row['name'])
        if row['label'] == 'career':
            settlement[time].append(row['name'])
        if row['label'] == 'settlement':
            creation[time].append(row['name'])
        if row['label'] == 'competition':
            assembly[time].append(row['name'])
        if row['label'] == 'military':
            military[time].append(row['name'])
        
        all_[time].append(row['name'])

    else:
        all_[time].append(row['name'])


for i in range(1900, 2001):
    performance[i] = set(performance[i])
    education[i] = set(education[i])
    start_org[i] = set(start_org[i])
    career[i] = set(career[i])
    settlement[i] = set(settlement[i])
    creation[i] = set(creation[i])
    assembly[i] = set(assembly[i])
    military[i] = set(military[i])
    all_[i] = set(all_[i])

# print(all_[1990].difference(i8n[1990]))
# get_trajectory('Ren√© Bazinet')
for i in range(1900, 2001):
    performance[i] = set(performance[i])
    education[i] = set(education[i])
    start_org[i] = set(start_org[i])
    career[i] = set(career[i])
    settlement[i] = set(settlement[i])
    creation[i] = set(creation[i])
    assembly[i] = set(assembly[i])
    military[i] = set(military[i])
    all_[i] = set(all_[i])
    
    # performance[i] = len(set(performance[i]))
    # education[i] = len(set(education[i]))
    # start_org[i] = len(set(start_org[i]))
    # career[i] = len(set(career[i]))
    # settlement[i] = len(set(settlement[i]))
    # creation[i] = len(set(creation[i]))
    # assembly[i] = len(set(assembly[i]))
    # military[i] = len(set(military[i]))
    # all_[i] = len(set(all_[i]))
    
    performance[i] = (len(set(performance[i]))) 
    education[i] = (len(set(education[i])) + performance[i]) 
    start_org[i] = (len(set(start_org[i])) + education[i]) 
    career[i] = (len(set(career[i])) + start_org[i]) 
    settlement[i] = (len(set(settlement[i])) + career[i]) 
    creation[i] = (len(set(creation[i])) + settlement[i])
    assembly[i] = (len(set(assembly[i])) + creation[i])
    military[i] = (len(set(military[i])) + assembly[i])
    
for i in range(1900, 2001):
    performance[i] = performance[i] / len(set(all_[i]))
    education[i]  = education[i]  / len(set(all_[i]))
    start_org[i]  = start_org[i]  / len(set(all_[i]))
    career[i] = career[i] / len(set(all_[i]))
    settlement[i] = settlement[i] / len(set(all_[i]))
    creation[i] = creation[i] / len(set(all_[i]))
    assembly[i] = assembly[i] / len(set(all_[i]))
    military[i] = military[i] / len(set(all_[i]))
```

```python

x = range(1900, 2001)
y_performance = [performance[i] for i in x]
y_education = [education[i] for i in x]
y_start_org = [start_org[i] for i in x]
y_career = [career[i] for i in x]
y_settlement = [settlement[i] for i in x]
y_creation = [creation[i] for i in x]
y_assembly = [assembly[i] for i in x]
y_military = [military[i] for i in x]

plt.figure(figsize=(16, 6))
colors = ["#FFD700",  # ÈáëËâ≤
    "#FFA500",  # Ê©ôËâ≤
    "#FF4500",  # Ê∑±Ê©ôËâ≤
    "#E91E63",  # Ê∑±Á≤âËâ≤
    "#C2185B",  # ÊöóÁ≤âËâ≤
    "#9C27B0",  # ‰∫ÆÁ¥´Ëâ≤
    "#945070",  # Ê∑±Á¥´Ëâ≤
    "blue"   # ÊöóÁ¥´Ëâ≤
]
colors.reverse()


plt.bar(x, y_military, label='military', color=colors[0])
plt.bar(x, y_assembly, label='competition', color=colors[1])
plt.bar(x, y_creation, label='settlement', color=colors[2])
plt.bar(x, y_settlement, label='career', color=colors[3])
plt.bar(x, y_career, label='movement', color=colors[4])
plt.bar(x, y_start_org, label='start org', color=colors[5])
plt.bar(x, y_education, label='education', color=colors[6])
plt.bar(x, y_performance, label='give birth', color=colors[7])

# plt.yscale('log')
# plt.title('Germany')
plt.legend(fontsize=16, ncol=2, loc='upper left')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

if country == 'Germany':
    plt.tight_layout()
    plt.savefig('./Germany_dis_distri.pdf', format='pdf')
else:
    plt.ylim((0, 0.6))
    plt.tight_layout()
    plt.savefig('./US_dis_distri.svg', format='svg')
```
